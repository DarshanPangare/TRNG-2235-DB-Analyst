#### Three Types of Principals

1. **Users**  
   - Human users who access the Databricks platform via UI, API, or CLI.  
   - Typically assigned individual roles and permissions.

2. **Service Principals**  
   - Non-human identities used for automation, integrations, or external tools.  
   - Allow programmatic access with specific permissions.

3. **Account Groups**  
   - Logical collections of users and/or service principals.  
   - Simplifies permission management by assigning rights at group level rather than individually.


#### Object Ownership and Permissions

- Every object in Databricks (tables, views, notebooks, clusters, etc.) has an **owner**.  
- Owners control permissions on their objects and can delegate access.  
- Permissions define allowed actions such as **READ**, **WRITE**, **MANAGE**, **EXECUTE**.  
- Permissions can be granted at different scopes forming a hierarchy.

#### Permissions Hierarchy for Data Objects

- **Catalog** (top-level container for schemas and tables)  
- **Schema** (namespace for tables and views)  
- **Table/View** (data objects)  

- Permissions flow down the hierarchy but can be overridden at lower levels for fine-grained control.


#### Permissions on Non-Data Objects

Databricks also controls access to compute and workspace assets beyond data:

- **Notebooks**  
  - Permissions control who can **view**, **edit**, **run**, or **manage** notebooks.  
  - Owners can share notebooks selectively to control collaboration.

- **Clusters**  
  - Permissions include who can **create**, **start**, **restart**, **terminate**, or **manage** cluster configurations.  
  - Prevents unauthorized resource usage and cost overruns.

- **Jobs**  
  - Control who can **create**, **edit**, **run**, or **manage** scheduled jobs.  

- **Libraries**  
  - Permissions determine who can **install** or **manage** libraries on clusters.


#### Sharing and Grants

- Databricks supports **fine-grained access control** via grants to principals.  
- Permissions are explicit and auditable.  
- Unity Catalog enables data sharing across accounts and workspaces without duplicating data.  
- Shares allow consumers to query data securely with access controls managed by providers.


#### Principle of Least Privilege

- Always grant **only the minimum permissions necessary** for a user or service to perform their job.  
- Reduces risk of accidental or malicious data exposure or system misuse.  
- Regularly review and audit permissions to ensure compliance with least privilege.
